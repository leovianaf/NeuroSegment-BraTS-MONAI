# NeuroSegment-BraTS: Modelagem e Experimentação

**Notebook 4:** Treinamento do Modelo e Avaliação por Métricas de Segmentação.

Nesta etapa, desenvolvemos a fase experimental do projeto, utilizando os dados tratados no pipeline de pré-processamento para treinar e avaliar um modelo de segmentação de tumores cerebrais. O foco deste notebook é documentar a configuração dos experimentos, os hiperparâmetros utilizados, os logs de treinamento e as métricas escolhidas para avaliação.

Com base no desbalanceamento observado na Análise Descritiva, a acurácia não será utilizada como métrica principal, pois pode favorecer modelos que classificam corretamente a classe majoritária sem segmentar adequadamente as regiões tumorais. Por isso, a avaliação será conduzida por métricas de sobreposição espacial, especialmente **Dice Score** e **Intersection over Union (IoU)**.

Essas métricas permitem medir a qualidade da segmentação comparando a máscara prevista pelo modelo com a máscara real, fornecendo uma avaliação mais adequada para o contexto de imagens médicas 3D.
